## Create a function to append to a new DF all the info pulled from Brickset website

```python
def content(x_list):
    df = pd.DataFrame()
    for x in x_list:
        df = df.append(pd.read_csv(f'https://brickset.com/exportscripts/inventory/{x}'))
    return df       
```
